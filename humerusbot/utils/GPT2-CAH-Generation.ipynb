```
sess = gpt2.start_tf_sess()
```

```
gpt2.finetune(
    sess,
    dataset="answers.txt",
    steps=1000,
    restore_from="fresh",
    run_name="white_cards",
    print_every=10,
    sample_every=200,
    save_every=500,
)
```

```
!cp -r /content/checkpoint/white_cards/ /content/drive/My\ Drive/UTMIST/
```

```
t = []
for i in range(100):
    t.append(gpt2.generate(sess, run_name='white_cards', top_k=50, temperature=0.8, return_as_list=True)[0])
```

```
o = [it for sl in [i.split('\n') for i in t] for it in sl]
```

```
!mkdir output
ofile = open("/content/output/white_cards.txt", 'a+')
for i in o:
    ofile.write(i+'\n')
ofile.close()
```
